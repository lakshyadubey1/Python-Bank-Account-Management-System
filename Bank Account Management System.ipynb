```python
import pickle
import numpy as np

# Load account data
try:
    with open("accounts.pkl", "rb") as file:
        accounts = pickle.load(file)
except FileNotFoundError:
    accounts = {}

# Load transaction data
try:
    with open("transactions.pkl", "rb") as file:
        transactions = pickle.load(file)
except FileNotFoundError:
    transactions = []


# Save all data
def save_data():

    with open("accounts.pkl", "wb") as file:
        pickle.dump(accounts, file)

    with open("transactions.pkl", "wb") as file:
        pickle.dump(transactions, file)


# Create account
def create_account():

    name = input("Enter Name: ")
    acc_no = len(accounts) + 1001

    print("Account Number:", acc_no)

    acc_type = input("Enter Account Type: ")
    balance = float(input("Enter Balance: "))

    accounts[acc_no] = {
        "name": name,
        "type": acc_type,
        "balance": balance
    }

    print("Account Created Successfully")


# View account
def view_account():

    acc_no = int(input("Enter Account Number: "))

    if acc_no in accounts:

        print("\nName :", accounts[acc_no]["name"])
        print("Type :", accounts[acc_no]["type"])
        print("Balance :", accounts[acc_no]["balance"])

    else:
        print("Account Not Found")


# Deposit money
def deposit():

    acc_no = int(input("Enter Account Number: "))

    if acc_no not in accounts:
        print("Account Not Found")
        return

    amount = float(input("Enter Amount: "))

    if amount <= 0:
        print("Invalid Amount")
        return

    accounts[acc_no]["balance"] += amount

    transactions.append({
        "account": acc_no,
        "type": "Deposit",
        "amount": amount
    })

    print("Deposit Successful")


# Withdraw money
def withdraw():

    acc_no = int(input("Enter Account Number: "))

    if acc_no not in accounts:
        print("Account Not Found")
        return

    amount = float(input("Enter Amount: "))

    if amount <= 0:
        print("Invalid Amount")

    elif amount > accounts[acc_no]["balance"]:
        print("Insufficient Balance")

    else:

        accounts[acc_no]["balance"] -= amount

        transactions.append({
            "account": acc_no,
            "type": "Withdrawal",
            "amount": amount
        })

        print("Withdrawal Successful")


# Transfer money
def transfer():

    sender = int(input("Sender Account Number: "))
    receiver = int(input("Receiver Account Number: "))

    if sender not in accounts or receiver not in accounts:
        print("Account Not Found")
        return

    amount = float(input("Enter Amount: "))

    if amount <= 0:
        print("Invalid Amount")

    elif amount > accounts[sender]["balance"]:
        print("Insufficient Balance")

    else:

        accounts[sender]["balance"] -= amount
        accounts[receiver]["balance"] += amount

        transactions.append({
            "account": sender,
            "type": "Transfer",
            "amount": amount
        })

        print("Transfer Successful")


# View transactions
def view_transactions():

    print("\nTransaction History")

    if len(transactions) == 0:
        print("No Transactions Found")

    else:
        for t in transactions:
            print(t)


# Generate report
def report():

    amounts = []

    for t in transactions:
        amounts.append(t["amount"])

    if len(amounts) == 0:
        print("No Transactions Found")

    else:
        print("\nTotal Transactions :", len(amounts))
        print("Total Amount :", np.sum(amounts))
        print("Average Amount :", np.mean(amounts))


# Main menu
while True:

    print("\n===== BANK SYSTEM =====")
    print("1. Create Account")
    print("2. View Account")
    print("3. Deposit")
    print("4. Withdraw")
    print("5. Transfer")
    print("6. View Transactions")
    print("7. Report")
    print("8. Exit")

    try:
        choice = int(input("Enter Choice: "))

        if choice == 1:
            create_account()

        elif choice == 2:
            view_account()

        elif choice == 3:
            deposit()

        elif choice == 4:
            withdraw()

        elif choice == 5:
            transfer()

        elif choice == 6:
            view_transactions()

        elif choice == 7:
            report()

        elif choice == 8:
            save_data()
            print("Data Saved")
            print("Thank You")
            break

        else:
            print("Invalid Choice")

    except ValueError:
        print("Enter Numbers Only")
```
